```python
df = pd.read_csv("./data/in/zillow_affordability.csv")
df.head()
```

```python
df = df[df["StateName"] == "CA"]
df.drop(columns=["RegionID", "SizeRank", "StateName", "RegionType"], inplace=True)
```

```python
df_a = df.melt(id_vars=["RegionName"], var_name="Date", value_name="Income Needed")
df_a.head()
```

```python
df_combined = df_zhvi.merge(df_a, on=["RegionName", "Date"], how="outer")
df_combined.head()
```

```python
avg_by_region = df_combined.groupby("RegionName").mean("Income Needed")
idx = avg_by_region.isna()["Income Needed"].values
cities_with_no_income_data = avg_by_region[idx].index.tolist()
```

```python
df = df_combined[~df_combined["RegionName"].isin(cities_with_no_income_data)]
df = df[~df["Income Needed"].isna()]
df.reset_index(inplace=True, drop=True)
df
```

```python
df.isna().sum().sum()
```

```python
avg_by_region = df.groupby("RegionName").count()
idx = avg_by_region["ZHVI"] != 157
avg_by_region[idx]
```

```python
df_la = df[df["RegionName"] == "Los Angeles, CA"]
df_la_bad_idx = df_la[df_la["ZHVI"].isna()].index.values[0]
df_la = df_la.reset_index()
df_la["ZHVI"] = df_la["ZHVI"].combine_first(
    (df_la["ZHVI"].ffill() + df_la["ZHVI"].bfill()) / 2
)
df.loc[df_la_bad_idx, "ZHVI"] = df_la[df_la["index"] == df_la_bad_idx]["ZHVI"].values[0]
```

```python
df_sj = df[df["RegionName"] == "San Jose, CA"]
df_sj_bad_idx = df_sj[df_sj["ZHVI"].isna()].index.values[0]
df_sj = df_sj.reset_index()
df_sj["ZHVI"] = df_sj["ZHVI"].combine_first(
    (df_sj["ZHVI"].ffill() + df_sj["ZHVI"].bfill()) / 2
)
df.loc[df_sj_bad_idx, "ZHVI"] = df_sj[df_sj["index"] == df_sj_bad_idx]["ZHVI"].values[0]
```

```python
df.isna().sum().sum()
```

```python
df
```

```python
df.drop(df[df["Date"].str.contains("2025")].index, inplace=True)
df.to_csv("./data/out/zillow.csv", index=False)
```

```python
df = pd.read_csv("./data/in/data.gov_personal_income.csv")
df.head()
```

```python
df.info()
```

```python
# df["Year"] = df["Year"].apply(lambda x: x.split("/")[-1].split()[0]).astype(int)
```

```python
tmpdf = df
tmpdf["Personal Income"] = tmpdf["Personal Income"].astype(str).str.replace("$", "\\$")
tmpdf.head()
```

```python
df["Age"].value_counts()
```

```python
df["Age"].replace(
    {
        "18 to 64": "18-64",
        "00 to 17": "<18",
        "65 to 80+": "65+",
    },
    inplace=True,
)
```

```python
df["Gender"].value_counts()
```

```python
df["Educational Attainment"].value_counts()
```

```python
df["Personal Income"].value_counts()
```

```python
df["income_numerical"] = df["Personal Income"]


def convert_income_numerical(value):
    if re.match(".*\d.*", value):
        lohi = value.split()
        lo, hi = lohi[0], lohi[-1]
        lo = int(str(lo).replace("\\", "").replace("$", "").replace(",", ""))
        if hi == "over":
            return lo
        else:
            hi = int(str(hi).replace("\\", "").replace("$", "").replace(",", ""))
            return (lo + hi) // 2
    else:
        return 0


df["income_numerical"] = df["income_numerical"].apply(convert_income_numerical)
```

```python
df["Population Count"] = df["Population Count"].fillna(0)
```

```python
df.info()
```

```python
df.rename(
    columns={
        "Year": "year",
        "Age": "age",
        "Gender": "gender",
        "Educational Attainment": "education",
        "Personal Income": "income_range",
        "Population Count": "count",
    },
    inplace=True,
)
```

```python
df_demographics = df[
    [
        "year",
        "age",
        "gender",
        "education",
        "income_range",
        "income_numerical",
        "count",
    ]
]
```

```python
df_main = pd.read_csv("./data/out/zillow.csv")
```

```python
df_main.info()
```

```python
df_main["Date"].value_counts()
```

```python
df_main["year"] = df_main["Date"].apply(lambda x: x.split("-")[0]).astype(int)
df_main.drop(columns=["Date", "RegionName"], inplace=True)
```

```python
df_avg = df_main.groupby("year").mean().reset_index()
df_max = df_main.groupby("year").max().reset_index()
df_min = df_main.groupby("year").min().reset_index()
df_med = df_main.groupby("year").median().reset_index()
```

```python
df_tmp = df_demographics
df_tmp = df_tmp.merge(df_avg, on="year", how="inner", suffixes=("", "_avg"))
df_tmp = df_tmp.merge(df_max, on="year", how="inner", suffixes=("", "_max"))
df_tmp = df_tmp.merge(df_min, on="year", how="inner", suffixes=("", "_min"))
df_tmp = df_tmp.merge(df_med, on="year", how="inner", suffixes=("", "_med"))
df_tmp.rename(
    columns={"ZHVI": "ZHVI_avg", "Income Needed": "Income Needed_avg"}, inplace=True
)
```

```python
df_tmp.info()
```

```python
df = df_tmp
df.drop(df[df["year"] == 2025].index, inplace=True)
df.to_csv("./data/out/zillow+demographics.csv", index=False)
```
